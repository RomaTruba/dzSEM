Задание 1.Использовать facebook/bart-large-cnn с другими языками используя обходные пути, такие как переводчики. Найдите примеры их использования на HuggingFace, например, facebook/m2m100_418M. Напишите примеры промтов для задачи суммаризации текста на русском и выведите ответ также на русском.

```python
from transformers import pipeline
```

```python
translator_ru_en = pipeline("translation", model="Helsinki-NLP/opus-mt-ru-en")
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
translator_en_ru = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ru")
```

```python
text_ru = """
Москва — столица Российской Федерации, крупнейший по численности населения город России и её субъект — 13 010 96 из крупнейших в Европе с населением более 20 млн человек.

Расположен в центре Восточно-Европейской равнины, в междуречье Оки и Волги, на стыке Смоленско-Московской возвышенности (на западе), Москворецко-Окской равнины (на юго-востоке) и Мещёрской низменности (на востоке). Город пересекает река Москва и её притоки.
"""
```

```python
text_en = translator_ru_en(text_ru)[0]['translation_text']
print(f"Текст на английском:\n{text_en}\n")
```

```python
summary_en = summarizer(text_en, max_length=130, min_length=30, do_sample=False)[0]['summary_text']
print(f"Суммаризация на английском:\n{summary_en}\n")
```

```python
summary_ru = translator_en_ru(summary_en)[0]['translation_text']
print(f"Суммаризация на русском:\n{summary_ru}")
```

Задание 2.Использовать другие модели для перевода и для задачи суммаризации текста, такие как T5, GPT-2, Pegasus и другие. Сравните результаты.

```python



summarizer = pipeline("summarization", model="google/pegasus-large")


text = """
Moscow is the capital of Russia. It is known for the Red Square and the Kremlin.
"""


summary = summarizer(text, max_length=130, min_length=30, do_sample=False)[0]['summary_text']
print(f"Суммаризация: {summary}")
```

Задание 3.Попробуйте составить промты для описания столиц мира, например, "Moscow is the capital of Russia. It is known for the Red Square and the Kremlin." Тексты можно взять с Википедии.

```python


generator = pipeline('text-generation', model='gpt2')  # Или другая модель

def generate_capital_description(city, country):
    prompt = f"""
    {city} is the capital of {country}. It is known for
    """
    description = generator(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']
    return description

city = "Moscow"
country = "Russia"
description = generate_capital_description(city, country)
print(description)
```
